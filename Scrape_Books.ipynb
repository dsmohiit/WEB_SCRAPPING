```python
import pandas as pd
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
data = []

for page in tqdm(range(1, 51)):
  link = "https://books.toscrape.com/catalogue/page-" + str(page) + ".html"
  res = requests.get(link)
  soup = BeautifulSoup(res.text, "lxml")

  for sp in soup.find_all("li", class_= "col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    title = sp.find_all("a")[-1].get("title")
    link = "https://books.toscrape.com/catalogue/" + sp.find_all("a")[-1].get("href")
    img_link = "https://books.toscrape.com/" + sp.find("img").get("src")
    rating = sp.find("p").get("class")[1]
    price = sp.find("p", class_= "price_color").text
    avail = sp.find("p", class_= "instock availability").text.strip()

    data.append([title, link, img_link, rating, price, avail])
```

```python
data
```

```python
df = pd.DataFrame(data, columns= ["Title", "Book Link", "Image Link", "Rating", "Price (Euro)", "Availability"])
```

```python
data = []
for link in tqdm(df["Book Link"]):
  res = requests.get(link)
  soup = BeautifulSoup(res.text, "lxml")

  typ = soup.find("ul", class_= "breadcrumb").find_all("a")[2].text
  stock = soup.find("p", class_= "instock availability").text.strip()[10:12]

  data.append([typ, stock])
```

```python
data
```

```python
temp_df = pd.DataFrame(data, columns= ["Genre", "No. of Units in Stock"]).reset_index()
```

```python
temp_df
```

```python
import pandas as pd

# Assuming df and temp_df are your DataFrames
books = pd.merge(df, temp_df, how= "inner", left_index= True, right_index= True)
```
